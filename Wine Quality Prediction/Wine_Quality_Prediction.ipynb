Importing the Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
#loading the dataset into pandas DataFrame
wine_dataset = pd.read_csv("/content/Wine Quality.csv")
```

```python
#Number of rows and columns in the data set
wine_dataset.shape
```

```python
#1st 5 rows of the DataFrame
wine_dataset.head()
```

```python
#checking any missing values in all columes of DataFrame
wine_dataset.isnull().sum()
```

Data Analysis and Visualization

```python
#statistical measure of the dataset
wine_dataset.describe()
```

```python
#number of values for each quality
sns.catplot(x='quality', data=wine_dataset, kind='count')
```

```python
# Volatile Acidity Vs Quality
plot = plt.figure(figsize=(5, 5))
sns.catplot(x='quality', y='volatile acidity', data=wine_dataset)
```

```python
#citric acid vs quality
plot = plt.figure(figsize=(5, 5))
sns.catplot(x='quality', y='citric acid', data=wine_dataset)
```

**Correlation**

> Postive Correlation

> Negative Correlation

```python
correlation = wine_dataset.corr()
```

```python
#constructing a heatmap to understand the corelation between columns
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Data Preprocessing

```python
#Separate the data and label
X = wine_dataset.drop('quality', axis=1)
```

```python
print(X)
```

Label Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

Train & Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

> Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

> Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Train Data Accuracy: ", train_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Test Data Accuracy: ", test_data_accuracy)
```
